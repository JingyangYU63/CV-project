# Data Preprocessing

```python
import os
import numpy as np
import pandas as pd

import torch
import torchvision
from torchvision.io import read_image
import torchvision.transforms as T
import matplotlib.pyplot as plt
```

```python
# Folder Path
path = "/Users/ziruizhou/Desktop/all_images_17k/all_images_17k"
  
# Change the directory
os.chdir(path)
number_of_pic = len(os.listdir())

print('There are {} png pictures.'.format(number_of_pic))
```

```python
# transform the image into square size
transform = T.Resize((256,256))

# Save the images
Primary_Dataset = torch.empty(number_of_pic, 3, 256, 256, dtype = torch.uint8)

# Get all the files' name 
files_name = os.listdir()

# Read the shuffled class label
Class_Label = pd.read_csv("/Users/ziruizhou/Desktop/all_images_17k/all_images_17k_shuffled.csv")
Files = pd.DataFrame(files_name, columns = ['name'])

assert len(files_name) == Class_Label.shape[0]

# Inner join on the primary data
joined_file_and_label = Files.merge(Class_Label, how = 'inner', on = 'name')

# Assert we kept the correct sequence
assert np.all(joined_file_and_label['name'] == files_name), 'wrong sequence'

# Save the class label in a 1-dim tensor
label = torch.tensor(joined_file_and_label['class'].values)
```

```python
# Save all the images into Primary_Dataset
for i in range(number_of_pic):
    img = torchvision.io.read_image(files_name[i])
    resized_img = transform(img)
    
    # Assert we have the correct image size
    assert resized_img.size() == torch.Size([3, 256, 256]), 'Wrong dim'
    Primary_Dataset[i] = resized_img
```
